```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#Import the required libraries
```

```python
# Load the dataset
df = pd.read_csv('Prasanth- Competitors - ElectricCarData_Clean.csv')
```

```python
print(df.head())
```

```python
print(df.shape)
```

```python
print(df.info())
```

```python
print(df.describe())
```

```python
print(df.isnull().sum())
```

```python
plt.hist(df['PriceEuro'], bins=10)
plt.xlabel('Price (Euro)')
plt.ylabel('Frequency')
plt.title('Distribution of Car Prices')
plt.show()
```

```python
plt.scatter(df['Range_Km'], df['Efficiency_WhKm'])
plt.xlabel('Range (km)')
plt.ylabel('Efficiency (kWh/km)')
plt.title('Range vs. Efficiency')
plt.show()
```

```python
sns.countplot(x='Brand', data=df)
plt.xlabel('Brand')
plt.ylabel('Count')
plt.title('Number of Cars by Brand')
plt.xticks(rotation=90)
plt.show()
```

```python
corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
#Boxplot by segments
sns.boxplot(x='Segment', y='PriceEuro', data=df)
plt.xlabel('Segments')
plt.ylabel('Price (Euro)')
plt.title('Price Distribution by Segment')
plt.xticks(rotation=90)
plt.show()
```

```python
#Bar plots of Categorical Variables
sns.countplot(x='PowerTrain', data=df)
plt.xlabel('PowerTrain')
plt.ylabel('Count')
plt.title('Number of Cars by PowerTrain')
plt.xticks(rotation=90)
plt.show()

sns.countplot(x='PlugType', data=df)
plt.xlabel('PlugType')
plt.ylabel('Count')
plt.title('Number of Cars by PlugType')
plt.xticks(rotation=90)
plt.show()
```

```python
#Scatter plot matrix
sns.pairplot(df[['AccelSec', 'TopSpeed_KmH', 'Range_Km', 'Efficiency_WhKm']])
plt.title('Scatter Plot Matrix')
plt.show()
```

```python
#Violin plot to compare distributions
sns.violinplot(x='BodyStyle', y='Range_Km', data=df)
plt.xlabel('BodyStyle')
plt.ylabel('Range (km)')
plt.title('Range Distribution by BodyStyle')
plt.xticks(rotation=90)
plt.show()
```

```python
#Distribution of Numeric Variables
sns.displot(df['AccelSec'], kde=True)
plt.xlabel('Acceleration (0-100 km/h)')
plt.ylabel('Density')
plt.title('Distribution of Acceleration')
plt.show()
```
